### DECLARE NEEDED IMPORTS AND FUNCTIONS TO THE PROCESS

````python
#### IMPORTS ####
import pandas as pd
import seaborn as sns
import numpy as np
import psycopg2
from datetime import datetime
from datetime import timedelta
from dateutil.parser import parse
import re
import glob #read files
import sqlalchemy as sa #conect DB

path = r'C:\Users\marcia.cunha\Desktop\Acessos\Cursos\python\atividade_usp\usp_python\gui_dataset' # use your path
all_files = glob.glob(path + "/*.txt")


# DEFINE FUNCTION TO CONECT TO REDSHIFT
def get_data(sql):
    """connects to RSift and returns dataframe"""
    connection_parameters = {
            'host': 'gfg-dwh-prod.dafiti.io',
            'port': '5439',
            'database': 'dftdwh',
            'user': 'etl',
            'password': ''
            }
    conn = psycopg2.connect(**connection_parameters)
    cur = conn.cursor()
    cur.execute(sql)
    columns = [item.name for item in cur.description]
    results = cur.fetchall()
    return pd.DataFrame(results, columns=columns)

# READ THE QUERY FILE AND TRANSFORM THE TEXT TO AVOID ERROR WHEN RUNNING
def get_query_from_file(file_name):
    # READ SQL QUERY FILE
    file = open(file_name)
    query = file.read();
    file.close();
    # REMOVE COMMENTS IN THE QUERY
    comment = re.compile('\-{2}(.+?)\\n')
    query = comment.sub(' ', query).replace('\n', ' ').replace('  ', ' ')
    return query
````

## DataSet Initialization

In the next 2 steps, load and create the dataframes with the initial data for analysis.

````python
# READ SQL QUERY FILE
query = get_query_from_file("sales_metrics_rate.sql")
# LOAD DATAFRAME FROM QUERY RESULT
daily_sales_per_country = get_data(query)
# create a new field of date as bigint and a new field of date as datetime object
year_month_day = daily_sales_per_country["iso_date"].to_numpy(dtype="str")
year_month_day = np.array(np.char.replace(year_month_day, "-", ""), dtype="int")
daily_sales_per_country["year_month_day"] = year_month_day
daily_sales_per_country["date_obj"] = [datetime.strptime(iso_date, '%Y-%m-%d') for iso_date in daily_sales_per_country["iso_date"].to_numpy(dtype="str")]
# show a sample of values
daily_sales_per_country.info()
daily_sales_per_country.head()
````

## Possible causes
#### 1. sale_order_seller and sale_order_seller_item

> this situation is particular for BR but can be a bridge to analyse AR, CL and CO.
***reverse engineer may be needed to understand how data is being received, maybe you just need to reprocess***

The product cost is calculated in table integration_layer.int_sale_order_seller_item and then propagated to business layer. During the process data from tables **store** and **seller** are merged and the mandatory in the flow is the store (BOB) as you can see in the flow bellow, it means that missing data in seller tables does not block the chain to load until the end.


```
     seller ------\
seller_item--------\ (secondary)
                    \
                    / --------------> sale_item---> fact_sales
  store_item-------/ (primary /
      store-------/    mandatory)
```

You may have some problem with some extraction from SAP, so you can analyse if there are many orders/items missing in seller tables related to orders/items in BOB. In case of BR the qty of orders/items in seller tables (SAP) that matchs with BOB orders/items has a bit difference because of data integration, business rule or anything else but it's not alarming as you can see in the chart bellow.

***Notice that the variation rarely is above 1%***

```
Glossary

seller_order_percent      - percentage of orders in BOB not found in int_sale_order_seller
seller_item_order_percent - percentage of orders in BOB not found in int_sale_order_seller_item
seller_item_percent       - percentage of items in BOB not found in int_sale_order_seller_item
```

````python
order_item_in_seller_tables_plot = debug_for_sales.query(f"country_code=='{contry_code_to_filter}' and date_obj < '{yesterday}'")\
[["year_month_day", "seller_order_percent", "seller_item_order_percent", "seller_item_percent"]]
order_item_in_seller_tables_plot.set_index("year_month_day", inplace=True)

order_item_in_seller_tables_graph = sns.lineplot(data=order_item_in_seller_tables_plot)
order_item_in_seller_tables_graph.figure.set_size_inches(14, 7)
order_item_in_seller_tables_graph.set_title("% of Order Item in BOB not found in seller tables (25 days without yesterday)", fontsize=15)
order_item_in_seller_tables_graph.set_ylabel("%", fontsize=15)
order_item_in_seller_tables_graph.set_xlabel("Day of Month", fontsize=15)
````
